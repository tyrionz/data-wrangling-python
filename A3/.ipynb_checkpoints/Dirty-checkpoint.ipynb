```python
import re
from contextlib import suppress


PATTERNS = [
        # 0) 1-12-1963 => 1963-12-01
        (re.compile(r'(\d{1,2})-(\d{1,2})-(\d{4})$'), '{2}-{1:0>2}-{0:0>2}'),
        # 1) 1789-7-14 => 1789-07-14
        (re.compile(r'(\d{4})-(\d{1,2})-(\d{1,2})$'), '{0}-{1:0>2}-{2:0>2}'),
]

def cleanup(date):
    
    with suppress(ValueError):
        return str(int(date))

    for pattern, formater in PATTERNS:
        match = pattern.match(date)
        if match is not None:
            return formater.format(*match.groups())
        
    return date

def fix_month(date):
    date = date.split('-')
    fixed = date[0]
    
    if int(date[1]) > 12:
        fixed = fixed + str('-') + date[2] + str('-') +  date[1]
    else:
        fixed = fixed + str('-') + date[1] + str('-') + date[2]
    return fixed

dirty['date_fix'] = dirty['date'].apply(lambda x : cleanup(x))
dirty['date_fix'] = dirty['date_fix'].apply(lambda x: fix_month(x))
```

```python
#fix the wrong item in order_item
def order_item_fix(meal,items,total_price):
    menu_qtt = re.findall(r'[0-9]+', items)
    menu_item = re.findall(r'[a-zA-Z&]+', items) 
    count = 0
  
    if meal == 'Breakfast':
        take_price = bprice
        meal_list = Breakfast
    if meal == 'Lunch':
        take_price = lprice
        meal_list = Lunch
    if meal == 'Dinner':
        take_price = dprice 
        meal_list = Dinner
        
    i = 0
    for item in menu_item:
        if item in meal_list:
            total_price = total_price - int(menu_qtt[i])*take_price[item]
            i+=1
        else:
            out_list = item
            count = int(menu_qtt[i])
            i+=1
                
    price_o_outlist = total_price/count
            
    for item, price in take_price.items():
        if abs(price_o_outlist - price) < 0.1:
            fix_out_list = item
    items = items.replace(out_list,fix_out_list)
    
    return items
```

```python
def order_item_check(meal,items):
    error = 0
    menu_qtt = re.findall(r'[a-zA-Z&]+', items)
    if meal == 'Breakfast':
        meal_list = Breakfast
    if meal == 'Lunch':
        meal_list = Lunch
    if meal == 'Dinner':
        meal_list = Dinner
    for item in menu_qtt:
        if item not in meal_list:
            error = 1
            break
    return error
```

```python
import networkx as nx
G = nx.Graph() 
nodes['node'] = nodes['node'].apply(lambda x:int(x))
G.add_nodes_from(list(nodes['node']))
edges = edges[['u','v','distance(m)','street type','speed(km/h)']]
edges_dis = edges[['u','v','distance(m)']]
edges_distance_tuple = [tuple(x) for x in edges_dis.values]
for start, end, length in edges_distance_tuple:
    # You can attach any attributes you want when adding the edge
    G.add_edge(int(start), int(end), length=length)
restaurants_csv['node'] = restaurants_csv[['branch_lat','branch_lon']].apply(lambda x: get_node(*x), axis=1)
NS = int(restaurants_csv[restaurants_csv['branch_code'] == 'NS']['node'][0])
TP = int(restaurants_csv[restaurants_csv['branch_code'] == 'TP']['node'][1])
BK = int(restaurants_csv[restaurants_csv['branch_code'] == 'BK']['node'][2])
```

```python
def distance(u,v):
    dist = 0.
    dist = nx.shortest_path_length(G, u, v,weight="length")/1000
    return round(dist,3)
```
